```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

training_data_nf = pd.read_json("../json/tagged_remote_sensing_2017_no_floats.json").sample(52, random_state=1).drop([66, 84])
training_data = pd.read_json("../json/tagged_remote_sensing_2017.json").sample(52, random_state=1).drop([66, 84])
```

```python
num = np.linspace(0,50,50)
fig = plt.figure(figsize=(6,4))
plot_2d = fig.add_subplot(111)

plot_2d.set_xlabel("Point")
plot_2d.set_ylabel("EVI")

plot_2d.plot(num, training_data["evi"], label="floats")
plot_2d.plot(num, training_data_nf["evi"], label="integers")

plot_2d.legend()

fig.savefig(f"pdf/evi_computations.pdf", 
            dpi=600,
            format="pdf", 
            facecolor="none", 
            alpha=0, 
            edgecolor="none", 
            bbox_inches="tight",
            orientation="portrait")
```
