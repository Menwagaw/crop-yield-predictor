```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
from sklearn import preprocessing

data = pd.read_json("../json/tagged_remote_sensing_2017.json", orient="records")
```

```python
for c in data.columns:
    if c == "type":
        continue
    fig = plt.figure(figsize=(6,4))
    plot = fig.add_subplot(111)
    plot.set_xlabel(f"{c.capitalize()}")
    plot.set_ylabel("Density")
    for t in ["city", "water", "cropland", "wood"]:
        data[data["type"] == t][c].hist(bins="auto", density=True, label=t, grid=False, alpha=0.75)
    plot.legend()
    fig.savefig(f"pdf/{c}_histogram.pdf", 
                dpi=600,
                format="pdf", 
                facecolor="none", 
                edgecolor="none", 
                bbox_inches="tight",
                orientation="portrait")
```

```python
for c in data.columns:
    if c != "type":
        for t in ["city", "water", "cropland", "wood"]:
            print(c, t, data[data["type"] == t][c].mean(), data[data["type"] == t][c].std())
```
